```python
# [1] Import Dependencies
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt
from tqdm import tqdm

# PyTorch Geometric imports
from torch_geometric.data import Data
from torch_geometric.nn import SAGEConv

# Hugging Face transformers for text and CLIP for images
from transformers import AutoTokenizer, AutoModel, CLIPProcessor, CLIPModel

# kgbench for the dataset
import kgbench as kg
```

```python
# [2] Load the Full Multimodal Dataset (final=False for rich literal data)
print("Loading dataset...")
data = kg.load('dmg777k', torch=True, final=False)
print("Loaded multimodal dataset:", data.name)
print("Number of entities:", data.num_entities)
print("Number of relations:", data.num_relations)
print("Number of classes:", data.num_classes)
```

```python
# [3] Create a Subset of the Data (if dataset is very large)
def create_subset(data, max_nodes=15000):
    if data.num_entities > max_nodes:
        print(f"Creating subset of {max_nodes} nodes...")
        subset_indices = np.random.choice(data.num_entities, max_nodes, replace=False)
        # Preserve edges where both subject and object are in the subset.
        mask = np.isin(data.triples[:, 0], subset_indices) & np.isin(data.triples[:, 2], subset_indices)
        subset_triples = data.triples[mask]
        # Create a mapping from old indices to new ones.
        node_map = {old: new for new, old in enumerate(subset_indices)}
        # Remap subject and object indices.
        ###########################################3remapped_triples = np.vectorize(node_map.get)(subset_triples)
        # Check if any node in subset_triples is missing in node_map:
                # Check if any node in subset_triples is missing in node_map:
        all_triple_nodes = set(subset_triples[:, 0]).union(set(subset_triples[:, 2]))
        missing_keys = all_triple_nodes - set(node_map.keys())
        if missing_keys:
            print("Warning: The following node indices are missing in node_map and will be removed:", missing_keys)
            # Create a mask that keeps only those triples where both subject and object are in node_map:
            mask = np.isin(subset_triples[:, 0], list(node_map.keys())) & np.isin(subset_triples[:, 2], list(node_map.keys()))
            subset_triples = subset_triples[mask]
            print("Filtered subset_triples shape:", subset_triples.shape)

        # Now remap the node indices using node_map. This version raises an error if a key is missing.
        ####################remapped_triples = np.vectorize(lambda x: node_map[x])(subset_triples)
        remapped_triples = np.vectorize(lambda x: node_map.get(x, -1))(subset_triples)
        # Remove triples with remapped indices of -1 (indicating missing nodes)
        remapped_triples = remapped_triples[(remapped_triples != -1).all(axis=1)]
        # Create a placeholder feature matrix for the subset (will be updated later)
        # And get labels for the subset.
        # Assume that data.training has labels for nodes (first column: node index, second: class)
        subset_labels = data.labels[subset_indices] if hasattr(data, 'labels') else None
        subset_data = Data(
            x=torch.randn(max_nodes, 128),  # Placeholder, will be replaced by extracted features.
            edge_index=torch.tensor(remapped_triples[:, [0, 2]].T, dtype=torch.long),
            y=torch.tensor(subset_labels) if subset_labels is not None else None
        )
        subset_data.num_nodes = max_nodes
        return subset_data, subset_indices, node_map
    else:
        return None, None, None

subset_data, subset_indices, node_map = create_subset(data, max_nodes=15000)
if subset_data is not None:
    print("Using subset data.")
else:
    print("Using full data for processing.")

# For further processing, we work on the full dataset if no subset is created.
# Here, we'll use the full dataset object for feature extraction and then later remap if using the subset.
working_data = data if subset_data is None else subset_data
```

```python
# [4] Set Up Hugging Face Models for Text and CLIP for Images
# Text Model (DistilBERT)
text_tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
text_model = AutoModel.from_pretrained("distilbert-base-uncased")
text_model.eval()

# CLIP Model for images (using CLIP ViT-B/32)
clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")
clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
clip_model.eval()
```

```python
# [5] Define Feature Extraction Functions
def extract_text_embedding(text):
    """Extract a 768-dimensional text embedding using DistilBERT."""
    inputs = text_tokenizer(text, return_tensors="pt", truncation=True, max_length=128)
    with torch.no_grad():
        outputs = text_model(**inputs)
    # Use mean pooling of token embeddings.
    embedding = outputs.last_hidden_state.mean(dim=1)  # shape: (1, 768)
    return embedding.squeeze(0).numpy()

def extract_image_embedding(image):
    """Extract a 512-dim image embedding using CLIP and project it to 128-dim."""
    inputs = clip_processor(images=image, return_tensors="pt")
    with torch.no_grad():
        outputs = clip_model.get_image_features(**inputs)  # shape: (1, 512)
    image_embedding = outputs.squeeze(0)  # shape: (512,)
    # Project image embedding to 128 dimensions using a simple linear layer.
    # For demonstration, we initialize a projection layer and use it in eval mode.
    proj = nn.Linear(512, 128)
    proj.eval()
    with torch.no_grad():
        projected = proj(image_embedding.unsqueeze(0)).squeeze(0).numpy()  # shape: (128,)
    return projected
```

```python
# [10] Define a Simple GNN Model (GraphSAGE based)
class MultiModalGNN(nn.Module):
    def __init__(self, in_dim, hidden_dim, out_dim):
        super(MultiModalGNN, self).__init__()
        self.conv1 = SAGEConv(in_dim, hidden_dim)
        self.conv2 = SAGEConv(hidden_dim, out_dim)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = F.relu(self.conv1(x, edge_index))
        x = F.dropout(x, p=0.5, training=self.training)
        x = self.conv2(x, edge_index)
        return F.log_softmax(x, dim=1)

model = MultiModalGNN(in_dim=working_data.x.shape[1], hidden_dim=256, out_dim=data.num_classes)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)
working_data = working_data.to(device)
```
