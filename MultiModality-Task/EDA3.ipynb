```python
from collections import Counter
print("Loaded multimodal dataset:", data.name)
print("Number of entities (nodes):", data.num_entities)
print("Number of relations:", data.num_relations)
print("Number of classes:", data.num_classes)

# --- Extract Node Information using the i2e Mapping ---
# data.i2e maps an integer index to a node representation (a pair: (literal, annotation)).
# For example, an element might be: ('"Some descriptive text"', '@nl-nl')
# or it could be ('', 'none') indicating an IRI or untagged literal.
node_reps = data.i2e  # This should be a list-like object of length data.num_entities

# Check the first few node representations.
print("\nSample node representations (first 5):")
for i in range(5):
    print(f"Node {i}: {node_reps[i]}")

# --- Count the Annotations ---
# Extract the annotation part from each node representation.
annotations = [ann for (literal, ann) in node_reps]
annotation_counts = Counter(annotations)
print("\nNode annotation counts:")
for ann, count in annotation_counts.items():
    print(f"{ann}: {count}")

# --- Visualize the Distribution of Node Annotation Types ---
plt.figure(figsize=(8,5))
plt.bar(annotation_counts.keys(), annotation_counts.values(), color='skyblue')
plt.xlabel("Node Annotation Type")
plt.ylabel("Count")
plt.title("Distribution of Node Annotation Types")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# --- Additional Analysis ---
# Optionally, you can also try to infer modality details from the literal part.
# For example, if the literal contains a URL or a specific pattern, you might tag it.
def infer_modality(literal, annotation):
    # A simple heuristic: if annotation is "none", we might treat it as an IRI.
    # If annotation starts with '@', it's a text literal.
    # Otherwise, you might check if the literal is a string that represents an image, etc.
    if annotation == "none":
        return "untagged"
    elif annotation.startswith("@"):
        return "Text literal"
    elif annotation=="http://kgbench.info/dt#base64image":
        return "image"
    # Extend this function if there are more specific datatype annotations.
    return annotation

modalities = [infer_modality(literal, ann) for (literal, ann) in node_reps]
modality_counts = Counter(modalities)
print("\nInferred node modality counts:")
for mod, count in modality_counts.items():
    print(f"{mod}: {count}")

plt.figure(figsize=(8,5))
plt.bar(modality_counts.keys(), modality_counts.values(), color='orange')
plt.xlabel("Inferred Node Modality")
plt.ylabel("Count")
plt.title("Distribution of Inferred Node Modalities")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
print("Loaded multimodal dataset:", data.name)
print("Number of entities:", data.num_entities)
print("Number of relations:", data.num_relations)
print("Number of classes:", data.num_classes)

# --- Create a DataFrame from the i2e Mapping ---
# data.i2e maps each node index to a tuple: (literal, annotation)
# We create a DataFrame with columns: "Index", "Literal", "Annotation"
node_list = []
for idx in range(data.num_entities):
    # If a node is missing from the mapping, you might need to handle that; here we assume it's present.
    literal, annotation = data.i2e[idx]
    node_list.append((idx, literal, annotation))

node_df = pd.DataFrame(node_list, columns=["Index", "Literal", "Annotation"])
print("\nFirst 5 nodes (from i2e mapping):")
print(node_df.head())

# --- (1) Count and Plot the Distribution of Annotations ---
annotation_counts = node_df["Annotation"].value_counts()
print("\nNode Annotation Counts:")
print(annotation_counts)

# Plot using pandas built-in plotting
annotation_counts.plot(kind="bar", color="skyblue", figsize=(8,5))
plt.xlabel("Annotation Type")
plt.ylabel("Count")
plt.title("Distribution of Node Annotations")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# --- (2) Infer Modalities Using a Simple Heuristic and Add as a New Column ---
def infer_modality(literal, annotation):
    # Simple heuristic:
    # - If annotation is "none", treat it as an IRI or untagged literal.
    # - If annotation starts with '@', treat it as a text literal.
    # - Otherwise, return the annotation as is.
    if annotation == "none":
        return "IRI/untagged"
    elif annotation.startswith("@"):
        return "Text literal"
    else:
        return annotation

# Apply the function to create a new "Modality" column.
node_df["Modality"] = node_df.apply(lambda row: infer_modality(row["Literal"], row["Annotation"]), axis=1)
print("\nFirst 5 nodes with inferred modality:")
print(node_df.head())

# Count modalities using value_counts on the new column.
modality_counts = node_df["Modality"].value_counts()
print("\nInferred Node Modality Counts:")
print(modality_counts)

# Plot modality distribution
modality_counts.plot(kind="bar", color="orange", figsize=(8,5))
plt.xlabel("Inferred Node Modality")
plt.ylabel("Count")
plt.title("Distribution of Inferred Node Modalities")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

##network graph:

```python
# --- Import Libraries ---
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx

print("Loaded multimodal dataset:", data.name)
print("Number of entities:", data.num_entities)
print("Number of relations:", data.num_relations)

# --- Convert Triples to a DataFrame ---
# data.triples is an (m, 3) array of (Subject, Relation, Object)
triples = data.triples.cpu().numpy() if hasattr(data.triples, 'cpu') else np.array(data.triples)
triples_df = pd.DataFrame(triples, columns=["Subject", "Relation", "Object"])
print("\nTotal triples:", triples_df.shape[0])
print("First 5 triples:")
print(triples_df.head())

# --- Create a Subgraph using a Subset of Triples ---
# For visualization, select the first 100 triples.
sub_triples_df = triples_df.head(100)

# Create a directed graph
G = nx.DiGraph()

# --- Add Nodes with Attributes (using data.i2e) ---
# Get all nodes appearing in the selected subset.
nodes_in_sub = set(sub_triples_df["Subject"]).union(set(sub_triples_df["Object"]))

# Define a simple function to choose a color based on annotation.
def choose_color(annotation):
    if annotation == "none":
        return "red"        # Red for nodes with no explicit annotation (likely IRIs/untagged)
    elif annotation.startswith("@"):
        return "blue"       # Blue for nodes with language tags (text literals)
    else:
        return "green"      # Green for any other annotation

# Add each node to the graph with its literal and annotation from data.i2e.
for node in nodes_in_sub:
    literal, annotation = data.i2e[node]
    color = choose_color(annotation)
    G.add_node(node, literal=literal, annotation=annotation, color=color)

# --- Add Edges with Relation Labels ---
for _, row in sub_triples_df.iterrows():
    subj = row["Subject"]
    obj = row["Object"]
    rel_idx = row["Relation"]
    # Convert relation index to human-readable relation using data.i2r.
    relation_str = data.i2r[rel_idx] if rel_idx < len(data.i2r) else "unknown"
    G.add_edge(subj, obj, relation=relation_str)

# --- Visualize the Graph ---
# Create a list of node colors from the node attributes.
node_colors = [G.nodes[node]["color"] for node in G.nodes()]

plt.figure(figsize=(12, 8))
# Use a spring layout for visualization.
pos = nx.spring_layout(G, seed=42)

# Draw nodes, edges, and labels.
nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=300)
nx.draw_networkx_edges(G, pos, arrowstyle="->", arrowsize=10)
nx.draw_networkx_labels(G, pos, font_size=8)

# Optionally, draw edge labels (showing relation types).
edge_labels = nx.get_edge_attributes(G, 'relation')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=6)

plt.title("Subgraph Representation of dmg777k with Distinct Node & Edge Types")
plt.axis('off')
plt.tight_layout()
plt.show()
```
